## **Time Series Analysis:**

```python
date=df['Date']
print(date)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df_indexed = df.set_index('Date')
monthly_sales = df_indexed['Total Amount'].resample('ME').sum()
plt.figure(figsize=(12,6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_sales_ma = monthly_sales.rolling(window=3).mean()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales, label='Original Sales')
plt.plot(monthly_sales_ma, label='3-Month Moving Average', color='red')
plt.title('Monthly Sales with Moving Average')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
daily_sales = df.set_index('Date')['Total Amount'].resample('ME').sum()

plt.figure(figsize=(12,6))
plt.plot(daily_sales)
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
monthly_growth = monthly_sales.pct_change() * 100

plt.figure(figsize=(12,6))
monthly_growth.plot(marker='o', color='purple')
plt.title('Month-on-Month Sales Growth (%)')
plt.xlabel('Month')
plt.ylabel('Growth Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The time series analysis of the retail sales data reveals clear sales trends and seasonal patterns. Monthly sales trends indicate a generally increasing trajectory, with some fluctuations suggesting varying demand across different periods. Analyzing sales by day of the week shows that weekends, particularly Saturdays and Sundays, tend to generate higher revenue, highlighting peak customer engagement during leisure periods. Daily sales trends reinforce this, showcasing visible spikes aligned with the weekends. Furthermore, the month-on-month growth rate helps identify periods of rapid expansion or decline, offering crucial insight into business performance over time. Seasonal decomposition reveals underlying components such as long-term trend, seasonal cycles, and irregularities in sales, providing a deeper understanding of consumer behavior patterns. Together, these time-based insights help identify optimal time frames for marketing strategies, stock planning, and forecasting.

# **Recommendations based on EDA:**

Based on the exploratory data analysis of the retail sales dataset, the following recommendations are provided to help the business make informed decisions and improve performance:

1.  **Target Age Group for Marketing:**
    *   **Insight:** The analysis of age distribution and total amount spent by age group indicates variations in purchasing power and behavior across different age segments. Identify the age groups with the highest total spending and quantity purchased.
    *   **Action:** Tailor marketing campaigns, product promotions, and loyalty programs to resonate with the most valuable age segments. Consider age-specific product recommendations and advertising channels.

2.  **Optimize Inventory and Promotions for High-Performing Categories:**
    *   **Insight:** "Clothing" and "Electronics" are the top-performing product categories in terms of quantity sold and revenue generated.
    *   **Action:** Ensure sufficient stock levels for popular items in these categories. Develop targeted promotions, bundle offers, and cross-selling strategies for related products within these categories to increase average transaction value.

3.  **Implement Gender-Based Personalization:**
    *   **Insight:** The analysis shows slight differences in purchasing behavior between genders and potential preferences for certain product categories (e.g., Beauty and Clothing for females, Electronics for males).
    *   **Action:** Personalize the online and in-store shopping experience based on gender where appropriate. This could include tailored product recommendations, personalized marketing messages, and curated product displays.

4.  **Capitalize on Year-End Sales Spike:**
    *   **Insight:** The time series analysis clearly shows a significant increase in sales during the year-end months (October to December), likely due to festive seasons and holidays.
    *   **Action:** Plan inventory well in advance to meet the surge in demand during this period. Launch targeted marketing campaigns and special promotions leading up to and during these months to maximize sales revenue.

5.  **Encourage Bulk Purchases:**
    *   **Insight:** The analysis of quantity purchased per transaction suggests opportunities to encourage customers to buy more items.
    *   **Action:** Implement strategies such as volume-based discounts ("Buy 3, Get 1 Free"), tiered pricing, or bundled product offers to incentivize customers to increase the quantity of items purchased in a single transaction.

6.  **Leverage Price and Quantity Relationship:**
    *   **Insight:** The correlation analysis (heatmap) indicates a positive correlation between 'Quantity' and 'Total Amount', and a strong positive correlation between 'Price per Unit' and 'Total Amount'.
    *   **Action:** While increasing quantity is beneficial, also consider strategies related to pricing. For higher-priced items, focus on highlighting value and quality. For lower-priced items, volume-based strategies can be particularly effective. Analyze the relationship between price, quantity, and total amount for different product categories to fine-tune pricing and promotional strategies.

7.  **Further Explore Customer Behavior:**
    *   **Insight:** The current analysis provides a good overview, but deeper dives into customer segmentation based on purchasing patterns, lifetime value, and recency/frequency of purchases could yield further valuable insights.
    *   **Action:** Conduct further analysis to identify different customer segments and develop targeted strategies for each segment. Implement customer relationship management (CRM) initiatives to foster loyalty and encourage repeat purchases.
